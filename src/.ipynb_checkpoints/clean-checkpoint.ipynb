```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
# read in data 
student_mat = pd.read_csv("../data/raw_data/student/student-mat.csv", sep = ";", header=0)
student_por = pd.read_csv("../data/raw_data/student/student-por.csv", sep = ";", header=0)
```

```python
# merge data
student_final = pd.merge(student_mat, student_por, on = ["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])

# print out merged data
student_final.head(10)
```

```python
# output cleaned data
student_final.to_csv("../data/cleaned_data/cleaned.csv")
```
